```python
# Applying Transformation 1 - Filter Valid Deliveries from ball_by_ball_df | Filtered 8196 Records.

from pyspark.sql.functions import col

vdlvry_bbb_df = ball_by_ball_df.filter(col('extra_type') == 'No Extras')
vdlvry_bbb_df.display()
```

Transformation 2 = Calculate Total and Average Runs scored in each match and also innings.


```python
from pyspark.sql.functions import sum, avg

total_and_avg_runs = vdlvry_bbb_df.groupBy(col('match_id'),col('innings_no')).agg(sum('runs_scored').alias('Total_runs_Scored'), avg('runs_scored').alias('Avg_runs_scored'))

tot_and_avg_runs_sorted = total_and_avg_runs.orderBy(col('match_id'), ascending = True)

display(tot_and_avg_runs_sorted)
```

```python
from pyspark.sql.window import Window

windowSpec = Window.partitionBy(col('match_id'), col('innings_no')).orderBy(col('over_id'))

vdlvry_bbb_df = vdlvry_bbb_df.withColumn('Sum_of_running_total', sum('runs_scored').over(windowSpec))
vdlvry_bbb_df.display() 
```

Create A High Impact Ball Column, Should include:
1. 6 or more runs on that ball including extras. (or)
2. Wicket on that ball.

```python
from pyspark.sql.functions import when, col

high_impact_ball = vdlvry_bbb_df.withColumn('High_Impact_Ball', when((col('runs_scored') + col('extra_runs') >= 6) | (col('bowler_wicket') == True), True).otherwise(False))

high_impact_ball.display()
```

```python
win_margin_df = match_df.withColumn('Win_Margin_Type', when(col('win_margin') > 100, 'Huge Margin')\
    .when((col('win_margin') > 50) & (col('win_margin') < 100), 'Medium Margin')\
        .otherwise('Small Margin'))

display(win_margin_df)
```

```python
toss_impact_df = win_margin_df.withColumn('Toss_Impact', when((col('toss_winner') == (col('match_winner'))), True).otherwise(False))

display(toss_impact_df)
```
